```python
import numpy as np
import matplotlib.pyplot as plt
```

### Passeio aleat√≥rio unidimensional, onde um caminhante parte da posi√ß√£o $ùë•=0$ e, a cada instante de tempo, d√° um passo de uma unidade para a direita com probabilidade $ùëù$, ou para a esquerda com probabilidade $1 ‚àí ùëù$. O passeio continua at√© completar $ùëÅ = 1000$ passos.

### Para $p=0.5$

```python
def random_walk(N, p):
    """
    Simula um passeio aleat√≥rio unidimensional.

    Par√¢metros:
    N: N√∫mero total de passos do passeio.
    p: Probabilidade de dar um passo para a direita (1-p para a esquerda).

    Retorna:
    np.ndarray: Array contendo a posi√ß√£o do caminhante a cada passo.
    """
    rng = np.random.default_rng()
    passos = rng.choice([-1, 1], size=N, p=[1-p, p])  
    posicao = np.cumsum(passos)  
    return posicao

def plot_random_walk(N, p):
    """
    Gera e plota um passeio aleat√≥rio unidimensional.

    par√¢metros:
    N: N√∫mero total de passos do passeio.
    p: Probabilidade de dar um passo para a direita (1-p para a esquerda).

    Retorna:
    None
    """
    plt.figure(figsize=(10, 8))
    posicao = random_walk(N, p)
    plt.plot(range(1, N+1), posicao, label=f'p = {p}',color='black')
    
    plt.xlabel('N√∫mero de Passos')
    plt.ylabel('Posi√ß√£o')
    plt.legend()
    plt.show()


N = 1000
p = 0.5


plot_random_walk(N, p)
```

```python
def random_walk(N, p, num_simulacoes=100):
    """
    Simula m√∫ltiplos passeios aleat√≥rios unidimensionais.
    
    Par√¢metros:
    N: N√∫mero total de passos do passeio.
    p: Probabilidade de dar um passo para a direita (1-p para a esquerda).
    num_simulacoes: N√∫mero de caminhantes para calcular estat√≠sticas.
    
    Retorna:
    np.ndarray: Matriz com posi√ß√µes dos caminhantes a cada passo.
    """
    rng = np.random.default_rng()
    passos = rng.choice([-1, 1], size=(num_simulacoes, N), p=[1-p, p])  
    posicoes = np.cumsum(passos, axis=1)  
    return posicoes

def calcular_estatisticas(N, p, num_simulacoes=100):
    """
    Calcula a m√©dia e o desvio padr√£o da posi√ß√£o dos caminhantes.
    
    Par√¢metros:
    N: N√∫mero total de passos.
    p: Probabilidade de dar um passo para a direita.
    num_simulacoes: N√∫mero de caminhantes simulados.
    
    Retorna:
    tuple: Arrays da m√©dia e do desvio padr√£o ao longo dos passos.
    """
    posicoes = random_walk(N, p, num_simulacoes)
    media = np.mean(posicoes, axis=0)
    desvio_padrao = np.std(posicoes, axis=0)
    return media, desvio_padrao

def plot_estatisticas(N, p, num_simulacoes=100):
    """
    Plota a m√©dia e o desvio padr√£o da posi√ß√£o dos caminhantes.
    """
    media, desvio_padrao = calcular_estatisticas(N, p, num_simulacoes)
    passos = np.arange(1, N+1)
    
    plt.figure(figsize=(12, 6))
    
    # Gr√°fico da m√©dia
    plt.subplot(1, 2, 1)
    plt.plot(passos, media, color='blue', label='M√©dia da posi√ß√£o')
    plt.xlabel('N√∫mero de Passos')
    plt.ylabel('M√©dia da Posi√ß√£o')
    plt.title('M√©dia da Posi√ß√£o')
    plt.legend()
    
    # Gr√°fico do desvio padr√£o
    plt.subplot(1, 2, 2)
    plt.plot(passos, desvio_padrao, color='red', label='Desvio Padr√£o')
    plt.xlabel('N√∫mero de Passos')
    plt.ylabel('Desvio Padr√£o')
    plt.title('Desvio Padr√£o')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Par√¢metros
total_passos = 1000
p = 0.5
simulacoes = 1000

# Executar e plotar
plot_estatisticas(total_passos, p, simulacoes)
```

### Podemos observar que ap√≥s rodar mais de 100 vezes o codigo para $p=0.5$, vemos que o deslocamento tende a retornar para a posi√ß√£o inicial.

### J√° para a probabilidade $p=0.6$ a posi√ß√£o tende a sempre crescer j√° que a probalidade maior √© sempre ir para a direita(positiva).

### Rodando o c√≥digo 100 vezes usando $p=0.5$ e $p=0.6$ e fazendo o gr√°fico para p conjunto de todos os passeios.

```python
def plot_random_walk_for_p_values(N, p1, p2, num_runs=100):
    """
    Gera e plota m√∫ltiplos passeios aleat√≥rios unidimensionais para os valores de p fornecidos.

    par√¢metros:
    N: N√∫mero total de passos do passeio.
    p1, p2: Valores de p para simular os passeios.
    num_runs: N√∫mero de simula√ß√µes a serem realizadas para cada valor de p.

    Retorna:
    None
    """
    plt.figure(figsize=(10, 8))

    
    for _ in range(num_runs):
        posicao_p1 = random_walk(N, p1)
        plt.plot(range(1, N+1), posicao_p1, color='red')  
    

    for _ in range(num_runs):
        posicao_p2 = random_walk(N, p2)
        plt.plot(range(1, N+1), posicao_p2, color='blue')  
    
    # Adicionando r√≥tulos e t√≠tulo
    plt.xlabel('N√∫mero de Passos')
    plt.ylabel('Posi√ß√£o')
    plt.title(r'C√≥digo rodado 100 vezes usando $p=0.5$ e $p=0.6$ ')
    
 
    
    plt.show()

# Par√¢metros
N = 1000  
p1 = 0.5  
p2 = 0.6  
num_runs = 100  

# Plotar m√∫ltiplos passeios aleat√≥rios para p1 e p2 no mesmo gr√°fico
plot_random_walk_for_p_values(N, p1, p2, num_runs)
```
