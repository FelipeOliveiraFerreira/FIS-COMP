```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(N, p):
    """
    Simula um passeio aleatório unidimensional.

    Parâmetros:
    N: Número total de passos do passeio.
    p: Probabilidade de dar um passo para a direita (1-p para a esquerda).

    Retorna:
    np.ndarray: Array contendo a posição do caminhante a cada passo.
    """
    rng = np.random.default_rng()
    steps = rng.choice([-1, 1], size=N, p=[1-p, p])  # Gera os passos com probabilidades p e 1-p
    position = np.cumsum(steps)  # Calcula a posição acumulada
    return position

def plot_random_walk(N, p):
    """
    Gera e plota um passeio aleatório unidimensional.

    parâmetros:
    N: Número total de passos do passeio.
    p: Probabilidade de dar um passo para a direita (1-p para a esquerda).

    Retorna:
    None
    """
    plt.figure(figsize=(10, 8))
    position = random_walk(N, p)
    plt.plot(range(1, N+1), position, label=f'p = {p}',color='black')
    
    plt.xlabel('Número de Passos')
    plt.ylabel('Posição')
    plt.title('Passeio Aleatório Unidimensional')
    plt.legend()
    plt.show()


N = 1000
p = 0.5

plot_random_walk(N, p)
```
